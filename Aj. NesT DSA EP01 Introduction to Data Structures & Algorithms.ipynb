##What is an Algorithm?

**Algorithm คือ** วิธีการ Solution หรือชุดคำสั่งที่กำหนดไว้อย่างดี เพื่อแก้ไขปัญหาเฉพาะอย่าง โดยใช้ชุดของ Input เพื่อให้ได้มาซึ่ง Output ที่ต้องการ

**ตัวอย่างเช่น** 

**An algorithm to add two numbers:**

1. Take two number inputs (1, 2)

2. Add numbers using the + operator (1 + 2)

3. Display the result (3)

![alt text](https://media.geeksforgeeks.org/wp-content/cdn-uploads/20191016135223/What-is-Algorithm_-1024x631.jpg)

Image: geeksforgeeks



##Qualities of a Good Algorithm

1. Input และ Output ควรกำหนดไว้อย่างแม่นยำ

2. แต่ละขั้นตอนใน Algorithm ควรมีความชัดเจนและไม่คลุมเครือ Step by Step

3. Algrithms ที่ใช้ควรมีประสิทธิภาพสูงสุดในบรรดาวิธีการต่าง ๆ ที่ใช้ในการแก้ปัญหา

4. Algorithm ไม่ควรยึดติดกับภาษาการเขียนโปรแกรมใดโปรแกรมหนึ่ง แต่ควรเขียน Algorithm ในลักษณะที่สามารถใช้ภาษาการเขียนโปรแกรมต่าง ๆ ได้

## Algorithm Examples

**Algorithm 1: Add two numbers entered by the user** 

บวกตัวเลขสองตัวที่ผู้ใช้ป้อน
```
Step 1: Start
Step 2: Declare variables num1, num2 and sum. 
Step 3: Read values num1 and num2. 
Step 4: Add num1 and num2 and assign the result to sum.
        sum = num1 + num2 
Step 5: Display sum 
Step 6: Stop

```

**Algorithm 2: Find the largest number among three numbers** 

หาตัวเลขที่มีค่ามากที่สุดในสามตัวเลขนี้

```
Step 1: Start
Step 2: Declare variables a,b and c.
Step 3: Read variables a,b and c.
Step 4: If a > b
           If a > c
              Display a is the largest number.
           Else
              Display c is the largest number.
        Else
           If b > c
              Display b is the largest number.
           Else
              Display c is the greatest number.  
Step 5: Stop
```

**Algorithm 3: Find the factorial of a number**

หาค่า Factorial ของตัวเลขที่กำหนด

```
Step 1: Start
Step 2: Declare variables n, factorial and i.
Step 3: Initialize variables
          factorial = 1
          i = 1
Step 4: Read value of n
Step 5: Repeat the steps until i = n
     5.1: factorial = factorial*i
     5.2: i = i+1
Step 6: Display factorial
Step 7: Stop
```









##What is Data Structures?

**Data Structures คือ** โครงสร้างข้อมูล ที่เก็บข้อมูลที่ใช้ในการจัดเก็บข้อมูล Store Data และจัดระเบียบข้อมูล Organize Data เป็นวิธีการจัดเรียงข้อมูล Arranging Data บนคอมพิวเตอร์เพื่อให้สามารถเข้าถึง Accessed และอัพเดท Updated การทำงานของระบบได้อย่างมีประสิทธิภาพ

![alt text](http://res.cloudinary.com/dyd911kmh/image/upload/c_scale,f_auto,q_auto:best,w_700/v1512740202/Template_2_oxrskq.png)

Image: DataCamp

**Primitve Data Strcutures** คือ ข้อมูลที่ถูกกำหนดคุณสมบัติไว้แล้ว เราไม่สามารถกำหนดเพิ่มหรือสร้างเพิ่มได้อีก 

--> Ex. Integer, Float, String, Boolean 

**Non-Primitive Data Structures** คือ
คือ ข้อมูลที่ยังไม่ถูกกำหนดคุณสมบัติ เราสามารถสร้างขึ้นมาใหม่และกำหนดคุณสมบัติของชนิดข้อมูลได้

***Built-in***

--> Ex. Array, List, Tuple, Dictionary, Set, File

***User-Defined***

**Linear Data Structures** หรือ โครงสร้างข้อมูลเชิงเส้น คือ โครงสร้างที่มีการจัดเก็บข้อมูลหรือความสัมพันธ์ของข้อมูลในลักษณะเรียงต่อเนื่องกันเป็นลำดับ

--> Ex. Array, List, Stack, Queue

**Non-Linear Data Structures** หรือ โครงสร้างข้อมูลแบบไม่เชิงเส้น คือ โครงสร้างที่มีการจัดเก็บข้อมูลหรือความสัมพันธ์กับข้อมูลอื่นได้หลายตัว

--> Ex. Graphs, Trees

##Why Learn Data Structures and Algorithms?
*   บริษัทใหญ่ ๆ จ้าง **Programmers** ที่มีความชำนาญเป็นพิเศษในการเพิ่มประสิทธิภาพ Algorithms  
*   **Programming** หรือการเขียนโปรแกรม เป็นเรื่องเกี่ยวกับ Data Structures & Algorithms
*   ***Data Structures ใช้เพื่อเก็บข้อมูล***
*   ***Algorithms ใช้เพื่อแก้ปัญหาโดยใช้ข้อมูลนั้น***
*   **Data Structures & Algorithms** จะดำเนินการแก้ไขปัญหาอย่างละเอียด และให้ข้อมูลเชิงลึกได้ว่าการใช้ Solutions แก้ไขในแต่ละปัญหามีประสิทธิภาพเพียงใด จะทำให้สามารถเลือกตัวเลือกที่ดีที่สุดได้

**ช่วยในเรื่อง** 
*   **Scalability** => Time & Size (Time Complexity)
*   **Memory** => Space & Storage (Space Complexity)
